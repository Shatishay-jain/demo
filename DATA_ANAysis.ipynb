```python
def funt(df):
    df['%impression']=(df['imp']/df['imp'].sum())*100
    df['%detailsPage']=(df['detailsFlag']/df['detailsFlag'].sum())*100
    df['%bookings'] = (df['bookingFlag']/df['bookingFlag'].sum()) * 100
    df['%realised'] = (df['realisedBookingFlag']/df['realisedBookingFlag'].sum()) * 100
    df['rr']= df['realisedBookingFlag']/df['bookingFlag']
    df['br']=df['bookingFlag']/df['detailsFlag']
    df['ctr']=df['detailsFlag']/df['imp'].sum()
    df=df.drop(['realisedBookingFlag', 'bookingFlag', 'detailsFlag', 'imp'],axis=1)
                
    return df
```

```python
def fum(df,c):
    y=df.groupby(pd.cut(df[c],np.arange(0,df[c].max()+5,5)))["realisedBookingFlag", "bookingFlag", "detailsFlag", "imp"].sum()
    y=funt(y)
    
    return y
```

```python
s_date=df.groupby(['searchdate'])[['realisedBookingFlag', 'bookingFlag', 'detailsFlag', 'imp']].sum()
s_date=funt(s_date)
s_date['UNIQUE_HOTELS_COUNT']=df.groupby(['searchdate'])[ 'hotelid'].nunique()
```

```python
rank=df.groupby('rank')[['realisedBookingFlag', 'bookingFlag', 'detailsFlag', 'imp']].sum()
rank=funt(rank)
rank["UNIQUE_HOTELS_COUNT"]=df.groupby(['rank'])[ 'hotelid'].nunique()
```
